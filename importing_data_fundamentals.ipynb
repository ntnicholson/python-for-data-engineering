# Connection String Setup

```python
import os
from dotenv import load_dotenv

load_dotenv()
#
user = os.getenv('DB_USER')
password = os.getenv('DB_PASS')
host = os.getenv('DB_HOST')
database_name = os.getenv('DB_NAME')

DB_URL = f'mysql+mysqlconnector://{user}:{password}@{host}/{database_name}'
```

## Import using SQLAlchemy

```python
from sqlalchemy import create_engine, text

engine = create_engine(DB_URL) #'mysql+mysqlconnector://user:password@host/database_name'
df = dict()
#test connection
try:
    connection = engine.connect()
    rs = connection.execute(text('SELECT * FROM employees'))
    df = pd.DataFrame(rs.fetchall()) #rs.fetchmany(size=5)
    df.columns = rs.keys()
    connection.close()
except Exception as e:
    print(e)

print(df)
```

## SQL using Pandas

```python
engine = create_engine(DB_URL) #'mysql+mysqlconnector://user:password@host/database_name'
df = pd.read_sql_query('SELECT * FROM employees', engine)
print(df.head())
```
