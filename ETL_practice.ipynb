```python
import pandas as pd
import sqlalchemy
```

# Extract Methods

```python
def extract(file_name):
    return pd.read_csv(file_name)
```

```python
def extract_from_parquet():
    raw_parquet_data = pd.read_parquet('raw_data.parquet', engine='fastparquet')
    return raw_parquet_data
```

```python
def extract_from_sql(connection_uri, query):
    # Create an engine, query data and return Data Frame
   db_engine = sqlalchemy.create_engine(connection_uri)
   return pd.read_sql(query, db_engine)
```

```python
def transform(data_frame):
    # Filter data include only subset of columns
    return data_frame.loc[:, ["Age", "BP"]]
```

```python
def load(data_frame, file_name):
    data_frame.to_csv(file_name)
```

```python
path = 'data_files/Heart_Disease_Prediction.csv'
extracted_data = extract(file_name=path)
#print(extracted_data)
transformed_data = transform(data_frame=extracted_data)
#print(transformed_data)
load(data_frame=transformed_data, file_name="data_files/load_data_test.csv")
```
